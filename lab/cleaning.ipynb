```python
import pandas as pd
```

```python
data = pd.read_csv('fx_data.csv')
```

```python
data.head()
```

```python
data.shape
```

```python
columns = ['Date']+[col for col in data.columns if col.split(' ')[-1].lower()=='close']
columns
```

```python
data = data[columns]
data.columns = ['date']+['_'.join(col.lower().split('_')[:-1]) for col in columns if col!='Date']
```

```python
data.head()
```

```python
data.columns
```

```python
data['date'] = pd.to_datetime(data['date'].apply(lambda d: d.split('+')[0]), format = '%Y-%m-%d %H:%M:%S')
data.sort_values(by='date', inplace = True)
```
